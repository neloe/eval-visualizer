```python
import pandas
```

```python
evals = pandas.read_csv('evals.csv')
evals
```

```python
semesters = set(evals['Semester'])
kelly_colors = [ '222222', 'F3C300', '875692', 'F38400', 'A1CAF1', 'BE0032', 'C2B280', '848482', '008856', 'E68FAC',  'F99379', '604E97', 'F6A600', 'B3446C', 'DCD300', '882D17', '8DB600', '654522', 'E25822', '2B3D26']
import seaborn as sns
sns.color_palette("Blues", n_colors=11)
# 0: 'F2F3F4',
# 6: 'BE0032',
sems = ['Spring 2016', 'Summer 2016', 'Fall 2016', 'Spring 2017', 'Summer 2017', 'Fall 2017','Spring 2018', 'Summer 2018', 'Fall 2018',  'Spring 2019','Summer 2019']
colormap = dict(zip(sems, sns.color_palette("Paired", len(sems))))
sems = ['Spring 2016', 'Summer 2016', 'Fall 2016', 'Spring 2017', 'Summer 2017', 'Fall 2017','Spring 2018', 'Summer 2018', 'Fall 2018',  'Spring 2019','Summer 2019']
colors = [colormap[x] for x in evals['Semester']]
school_mean = 'black'
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as style
import matplotlib.patches as mpatches
import seaborn as sns
sns.color_palette("GnBu_d", n_colors=11)

red_patch = mpatches.Patch(color='red', label='The red data')
sems = ['Spring 2016', 'Summer 2016', 'Fall 2016', 'Spring 2017', 'Summer 2017', 'Fall 2017','Spring 2018', 'Summer 2018', 'Fall 2018',  'Spring 2019','Summer 2019']
patches = [mpatches.Patch(color=colormap[x], label=x) for x in sems]
style.use('seaborn-poster')

labels = ['{} (n={:0>2d})'.format(*x) for x in zip(evals['Class'], evals['n'])]
spacing = [2*x for x in range(len(evals)-1, -1, -1)]


fig=plt.figure(figsize=(7, 12))
ax = plt.subplot(111)
ax.barh(spacing,evals['Instructor'], color=colors, tick_label=labels, height=1)
csis, = ax.plot(evals['CSIS'], spacing, color=school_mean)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), handles=patches+[csis], labels=sems + ['CSIS Mean'])

for i, v in enumerate(evals['Instructor'][::-1]):
    x = v+.1
    if evals['CSIS'][::-1][i] > v:
        x = evals['CSIS'][::-1][i]+.1
    ax.text(x, 2*i, str(v), horizontalalignment='left', verticalalignment='center')
plt.xlim((0.0, 4.5))
plt.title('Course Averages Compared to School Mean')
plt.savefig('evals.png', bbox_inches='tight')
```
